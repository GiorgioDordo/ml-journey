```python
import random

def guess_number(n):
  random_number = random.randrange(1,101)

  print(random_number)

  if n == random_number:
    return "Congratulations! You've guessed the number!"
  elif n < random_number:
    return "Too low! Try again."
  else:
    return "Too high! Try again."
  

guess_number(50)

# a simple number guessing game where the user inputs a number and the program generates a random number between 1 and 100, then compares the two numbers and provides feedback on whether the user's guess is too low, too high, or correct.
```
