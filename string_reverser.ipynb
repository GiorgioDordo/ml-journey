```python
def string_reverse(string):
    """
    Reverse a string.
    Using a Loop
    Using  the string concatenation left comes first and the right comes last.
    Example:
    Input: "Hello, World!"
    Output: "!dlroW ,olleH"
    So the first character "H" will be added to the empty string, then the second character "e" will be added to the left of "H", and so on until the entire string is reversed. Because reversed_string is updated in each iteration of the loop, it will contain the reversed string at the end of the loop.

    """

    reversed_string = ""
    for char in string:
        reversed_string = char + reversed_string
    return reversed_string

string_reverse("Hello, World!")
```
