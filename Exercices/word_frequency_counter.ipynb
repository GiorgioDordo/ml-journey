```python
def word_counter(text):
    """
    Counts the frequency of each word given a string of text.
    """

    word_count = {}
    words = text.split()

    for word in words:
        if word == word:
            word_count[word] = word_count.get(word, 0) + 1
    return word_count

word_counter("I love pizza and I love pasta")
```
